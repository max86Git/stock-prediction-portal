```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Type of plots

#### Line plots

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 35, 50]
```

```python
plt.plot(x,y)
plt.title('Simple line plot')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

#### Bar charts

```python
months = ["Jan", "Fév", "Mar", "Avr"]
revenue = [100, 200, 350, 485]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(months, revenue)
plt.title('Monthly Revenue Bar Chart')
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.show()
```

#### Histogram

```python
marks50 = np.random.randint(0, 100,50)
marks50
```

```python
plt.hist(marks50)
plt.xlabel("Marks")
plt.ylabel("Students")
plt.show()
```

```python
plt.hist(marks50, bins=100)
plt.xlabel("Marks")
plt.ylabel("Students")
plt.show()
```

## subplots & Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 30, 35, 60]
predicted_price = [8, 21, 32, 38, 75]
```

```python
plt.plot(days, original_price, "b-o", label="Original price")
# "b" est la couleur bleu, "b-o" est la couleur bleu et affiche les points
plt.plot(days, predicted_price, "r-o", label="Predicted price")
plt.legend()
plt.title("Plot with legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.show()
```

## subplot

```python
# First sublot - Original Price
plt.subplot(3, 1, 1)
plt.plot(days, original_price, "b-o", label="Original price")
plt.title("Original prices over time")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(days, predicted_price, "r-o", label="Predicted price")
plt.title("Original prices over time")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()

# différence entre les prix
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3)
plt.bar(days, price_difference, color="purple", label="Predicted price")
plt.title("Prices difference over time")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()
plt.tight_layout()
plt.show()
```

## Saving the figure

```python
plt.plot(days, original_price, "b-o", label="Original price")
# "b" est la couleur bleu, "b-o" est la couleur bleu et affiche les points
plt.plot(days, predicted_price, "r-o", label="Predicted price")
plt.legend()
plt.title("Plot with legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.savefig("example_predict.png", dpi=300)
```
